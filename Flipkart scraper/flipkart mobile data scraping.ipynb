```python
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
#this is the url we are working with
url = 'https://www.flipkart.com'
```

```python
#driver for operation
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
#addding timeout so that it won't force the work and scrape everything
timeout = 100
```

```python
#maximizing the window 
driver.maximize_window()

#getting the url
driver.get(url)

#letting driver wait for a time
driver.implicitly_wait(timeout)
driver.find_element_by_xpath('''/html/body/div[2]/div/div/button''').click()
driver.implicitly_wait(timeout)

#searching the keyword just trying to scrape one keyword can scrape more keywords too but it is out of aim 
driver.find_element_by_xpath('''//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input''').send_keys('apple phones')
driver.implicitly_wait(timeout)
driver.find_element_by_xpath(''' //*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button''').click()
```

```python
import time

#try except loop so that our script will run without errors 
try:
    titles =[]
    ratings = []
    prices = []
    links = []
    for x in range(1,12):
        print(f"scraping page number {x}\n")
        driver.find_element_by_xpath(f''' //*[@id="container"]/div/div[3]/div[1]/div[2]/div[26]/div/div/nav/a[{x}]''').click()
        driver.implicitly_wait(timeout)
        #sleep the loop so that operation get performed smoothly
        time.sleep(3)
        for x in range(2,26):
            title = driver.find_element_by_xpath(f'''//*[@id="container"]/div/div[3]/div[1]/div[2]/div[{x}]/div/div/div/a/div[2]/div[1]/div[1]''').text
            rating =  driver.find_element_by_xpath(f'''//*[@id="container"]/div/div[3]/div[1]/div[2]/div[{x}]/div/div/div/a/div[2]/div[1]/div[2]''').text
            price = driver.find_element_by_xpath(f'''//*[@id="container"]/div/div[3]/div[1]/div[2]/div[{x}]/div/div/div/a/div[2]/div[2]/div[1]/div[1]/div[1]''').text
            href = driver.find_element_by_xpath(f'''//*[@id="container"]/div/div[3]/div[1]/div[2]/div[{x}]/div/div/div/a''').get_attribute("href")
            print(title)
            titles.append(title)
            print(rating)
            ratings.append(rating)
            print(price)
            prices.append(price)
            print(href)
            links.append(href)
            print()
        print('-'*125)
        print()
except:
    pass
```

```python
import pandas as pd ,numpy as np 
```

```python
#zippling the list so that it'll form the tuple
zipped = list(zip(titles,ratings,prices,links))
zipped
```

```python
csv_file = pd.DataFrame(zipped,columns=['titles','ratings','prices','links'])
csv_file
```

```python
csv_file.to_csv('Apple_phones.csv')
```
