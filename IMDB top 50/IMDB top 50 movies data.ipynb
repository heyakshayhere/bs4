```python
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
#this is the url we are working with
url = 'https://www.imdb.com/search/keyword/?sort=user_rating,desc&mode=detail&page=1&title_type=tvMovie&ref_=kw_ref_typ'
```

```python
#driver for operation
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
#addding timeout so that it won't force the work and scrape everything
timeout = 100
```

```python
#maximizing the window 
driver.maximize_window()

#getting the url
driver.get(url)
```

```python
import time

#try except loop so that our script will run without errors 
try:
    movies =[]
    releases = []
    directors_and_stars = []
    
    for x in range(1,51):
        movie = driver.find_element_by_xpath(f'''//*[@id="main"]/div/div[2]/div[3]/div[{x}]/div[2]/h3/a''').text
        release = driver.find_element_by_xpath(f'''//*[@id="main"]/div/div[2]/div[3]/div[{x}]/div[2]/h3/span[2]''').text
        director_and_stars = driver.find_element_by_xpath(f'''//*[@id="main"]/div/div[2]/div[3]/div[{x}]/div[2]/p[3]''').text
        print(movie)
        movies.append(movie)
        print(release)
        releases.append(release)
        print(director_and_stars)
        directors_and_stars.append(director_and_stars)
        print()
except:
    pass
```

```python
import pandas as pd ,numpy as np 
```

```python
#zippling the list so that it'll form the tuple
zipped = list(zip(movies,releases,directors_and_stars))
zipped[:5]
```

```python
csv_file = pd.DataFrame(zipped,columns=['movie','release','director_and_stars'])
csv_file
```

```python
csv_file.to_csv('IMDB top 50 movies.csv')
```
