```python
import pandas as pd,requests,bs4,time
from datetime import datetime
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
def parse_webpage_api(url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:90.0) Gecko/20100101 Firefox/90.0"}
        api_request = requests.get(url, headers=headers, timeout=20)
        if api_request.status_code==200:
            data = api_request.json()
            return data
        else:
            print("Access Denied")
    except:
        return []
```

```python
def remove_esc_chars(text):
    return text.replace("\n", " ").replace("\t", " ").replace("\r", " ")
```

```python
SITE_NAME='Novonordisk'

DOMAIN = "https://www.novonordisk.com"

SITE_LINK="https://www.novonordisk.com/news-and-media/news-and-ir-materials.html"
```

for pagination we can change the limit value  

```python
article_list = []
while True:
    dummy_url = "https://www.novonordisk.com/bin/nncorp/news-search?searchtext=null&start=2000-01-01&end=2022-05-24&newspath=/content/dam/nncorp/global/en/investors/content-fragments/ir-materials&limit=10&currentresults=0&function=search&language=all&type=Annual%20reports,Investor%20presentations,5355,5356"
    temp_data = parse_webpage_api(dummy_url)
    len_elements = temp_data['data']['numberOfResults']
    url = f"https://www.novonordisk.com/bin/nncorp/news-search?searchtext=null&start=2000-01-01&end=2022-05-24&newspath=/content/dam/nncorp/global/en/investors/content-fragments/ir-materials&limit={len_elements}&currentresults=0&function=search&language=all&type=Annual%20reports,Investor%20presentations,5355,5356"    
    data = parse_webpage_api(url)
    elements = data['data']['resultBeanList']
    if data != None:
        for element in elements:
            title = element['title'].strip()
            link = f"https://www.novonordisk.com/content/nncorp/global/en/news-and-media/news-and-ir-materials/news-details.html?id={element['id']}"
            date = element['date']
            published_date = date.split("T")[0].strip()
            article_html = element['newsContent']
            site_soup = BeautifulSoup(article_html, "lxml")
            paragraphs = site_soup.findAll('p')
            content = []
            for para in paragraphs:
                text_p = remove_esc_chars(para.get_text().strip())
                content.append(text_p)
            text = " ".join(content)
            author = SITE_NAME
            thumbnail = "https://www.novonordisk.com/etc.clientlibs/nncorp/components/structure/page/clientlib/resources/favicon.ico"
            article = (published_date,title,text,link,thumbnail,author)
            article_list.append(article)
            print(published_date,title)
    break
```

```python
temp_df = pd.DataFrame(article_list,columns=['date','title','article','url','thumbnail','author'])
temp_df.head()
```

```python
#to csv
temp_df.to_csv('novonordisk news.csv')

#to json
temp_df.to_json('novonordisk news.json')
```
