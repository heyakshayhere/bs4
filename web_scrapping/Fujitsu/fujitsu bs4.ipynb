```python
from bs4 import BeautifulSoup
from lxml import etree
import requests
import colorama
from colorama import Fore
from datetime import datetime
import pandas as pd , numpy as np ,re

import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
#for formatting use this reference 
#https://www.geeksforgeeks.org/python-strftime-function/#:~:text=The%20strftime()%20function%20is,and%20returns%20the%20string%20representation.&text=Returns%20%3A%20It%20returns%20the%20string,the%20date%20or%20time%20object.
# creating the list of dates 
list_of_dates = []

try:
    #taking input from user
#     start_date = input("Enter start date (in YYYY MM DD) : " )
#     end_date = input("Enter end date (in YYYY MM DD) : " )
    start_date = '2021 1 1'
    end_date = '2022 4 19'

    #adding up condition 
    if start_date > end_date :
        print(Fore.RED+ '\nstart date must come before end date')

    #appending to the list includes both dates
    else:
        for d in pd.date_range(start_date, end_date,inclusive="both"):
            list_of_dates.append(d.strftime("%B %#d, %Y"))

#dealing with error/s    
except ValueError:
    print(Fore.RED +"\nAdd the dates in correct format (in YYYY MM DD)")
```

```python
list_of_dates
```

```python
dates = []
titles = []
links = []


for pagination in range(2022,2020,-1):
        #getting the urls
        url =  f'https://www.fujitsu.com/global/about/resources/news/press-releases/{pagination}/'

        #parsing
        source = requests.get(url)
        soup = BeautifulSoup(source.content,'lxml')
        
        all_containers = soup.findAll('a', {'class' :'news-content'})
        
        for container in all_containers:
                published_date = container.find("div").text              
                if published_date in list_of_dates:
                    dates.append(published_date)
                    title = container.find("li").text
                    titles.append(title)
                    link = container['href']
                    if re.findall('.pdf',link):
                        links.append(link)
                    else:
                        links.append("https://www.fujitsu.com/" + link)
                    
                
```

```python
zipped = list(zip(dates,titles,links))
zipped[:3]
```

```python
#creating the csv 
temp_csv = pd.DataFrame(zipped,columns=['date','title','link'])
temp_csv
```

```python
%%time
articles = []
thumbnails = []

for link_number in range(len(temp_csv)):
    
    #getting the url from data
    url = temp_csv['link'][link_number]
    
    #parsing
    source = requests.get(url)
    soup = BeautifulSoup(source.content,'lxml')
     
    try:    
        #creating container and fetching the data    
        all_containers = soup.findAll('div',{'class' :"section-inner news-area"})
        thumbnail_container = soup.findAll('div',{'class' :"lyt-img-b"})        
        
        art = []
        for container in all_containers:
            article = container.text
            art.append(article)
        
        if len(art) == 0 :
             articles.append("")
        else:
            articles.append(art[0])
                  
        thumb = []
        
        for img_container in thumbnail_container:
            thumbnail = img_container.find('img')['src']            
            thumb.append("https://www.fujitsu.com" + thumbnail)
        try:
            thumbnails.append(thumb[1])
        except:
            thumbnails.append("")
            
        print(f"scrapping done for url {link_number}\n")
        
    except:
        pass
```

```python
temp_csv['Article'] = articles
temp_csv['Thumbnails'] = thumbnails
temp_csv
```

```python
#to csv
temp_csv.to_csv('fujitsu news.csv')

#to json
temp_csv.to_json('fujitsu news.json')
```
