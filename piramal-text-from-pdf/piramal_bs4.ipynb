```python
import pandas as pd,requests,bs4,re,time,io
from PyPDF2 import PdfReader
from goose3 import Goose
from datetime import datetime
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
SITE_NAME='Piramal'

DOMAIN = "https://www.piramal.com"

SITE_LINK="https://www.piramal.com/news-media/press-release/"
```

```python
def remove_esc_chars(text):
    return text.replace("\n", " ").replace("\t", " ").replace("\r", " ")
```

```python
def parse_webpage_bs(search_url):
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:90.0) Gecko/20100101 Firefox/90.0"}
    try:
        site_request = requests.get(search_url, headers=headers, timeout=10)
    except requests.exceptions.RequestException as e:
        print(e)
        site_request = None
    if site_request != None and site_request.status_code==200:
        site_soup = bs4.BeautifulSoup(site_request.content, "lxml")
    else:
        site_soup = None
    return site_soup
```

```python
article_list = []
url = SITE_LINK
soup = parse_webpage_bs(url)
if soup != None:
    elements = soup.findAll('td',{'class' :'nseardis'})[:20]
    for element in elements:
        try:
            #title
            title = element.find('p').text.strip()

            #published_date,link
            date = element.find('div')
            published_date = date.text.split(" ")[1].strip()
            link = date.find('a')['href']

            #text
            r = requests.get(link)
            f = io.BytesIO(r.content)
            reader = PdfReader(f)
            number_of_pages = len(reader.pages)
            page = reader.pages[0]
            text = remove_esc_chars(page.extract_text()).strip()

            #thumbnail and author
            thumbnail = "https://www.piramal.com/wp-content/uploads/2020/05/logo.svg"
            author = SITE_NAME

            article = (published_date,title,text,link,thumbnail,author)
            article_list.append(article)
            print(published_date,title)
        except:
            pass
        
        print(str(len(elements)- elements.index(element) -1) + " pages remaining..",end="\r")
```

```python
temp_df = pd.DataFrame(article_list,columns=['date','title','article','url','thumbnail','author'])
temp_df
```

```python
#to csv
temp_df.to_csv('Piramal news.csv',index = False)

#to json
temp_df.to_json('Piramal news.json')
```
